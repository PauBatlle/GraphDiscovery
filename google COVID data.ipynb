```python
import pandas as pd
```

```python
index=pd.read_csv('google_index.csv')
index
```

```python
index[index['country_name']=='France']
```

```python
gov=pd.read_csv('google_govment.csv')
gov=gov[gov['location_key']=='FR'].drop(columns='location_key')
print(gov.columns)
gov=gov[['date', 'school_closing', 'workplace_closing', 'cancel_public_events',
       'restrictions_on_gatherings', 'public_transport_closing',
       'stay_at_home_requirements', 'restrictions_on_internal_movement',
       'income_support',
       'facial_coverings', 'vaccination_policy', 'stringency_index']]
gov
```

```python
hospital=pd.read_csv('google_hospital.csv')
hospital = hospital[hospital['location_key']=='FR'].drop(columns='location_key')
print(hospital.columns)
hospital=hospital[['date', 'new_hospitalized_patients', 'cumulative_hospitalized_patients',
       'current_hospitalized_patients', 'new_intensive_care_patients',
       'cumulative_intensive_care_patients', 'current_intensive_care_patients',]]
hospital
```

```python
vaccine=pd.read_csv('google_vaccination.csv')
vaccine = vaccine[vaccine['location_key']=='FR'].drop(columns='location_key')
vaccine=vaccine[['date', 'new_persons_vaccinated', 'cumulative_persons_vaccinated',
       'new_persons_fully_vaccinated', 'cumulative_persons_fully_vaccinated',
       'new_vaccine_doses_administered',
       'cumulative_vaccine_doses_administered']]
print(vaccine.columns)
vaccine
```

```python
vaccine.isna().sum()
```

```python
epidemiology=pd.read_csv('google_epidemiology.csv')
epidemiology = epidemiology[epidemiology['location_key']=='FR'].drop(columns='location_key')
epidemiology
```

```python
epidemiology.isna().sum()
```

```python
hospital_and_gov= hospital.join(gov.set_index('date',drop=True), on='date')
h_g_e=hospital_and_gov.join(epidemiology.set_index('date',drop=True), on='date')
full=h_g_e.join(vaccine.set_index('date',drop=True), on='date')
```

```python
full.dropna(axis=0).to_csv('france_covid_data.csv',index=False)
```
